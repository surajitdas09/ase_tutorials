### Part 02 <i><b>ase.atoms</b></i>

```python
from ase import Atoms
h2 = Atoms('H2', positions=[[0,0,0],[0,0,0.74]]); print(h2)
```

```python
zi = h2.get_atomic_numbers(); print(zi) # --- atomic numbers 
ai = h2.get_chemical_symbols(); print(ai) # --- atomic symbols
ri = h2.get_positions(); print(ri) # --- atomic positins
qi = h2.get_masses(); print(qi) # --- atomic masses
```

#### <i><b>set_positions</b></i>

```python
h2.set_positions([(0.,0.,0.),(0.,0.,-0.74)])
ri = h2.get_positions(); print(ri)
```

#### <i><b>translate</b></i>

```python
h2.translate([1.,0.,0.]) # --- translate 
ri = h2.get_positions(); print(ri)
```

#### <i><b>rotate</b></i>

```python
h2.rotate(90,'x',center='COM') # --- rotate 
ri = h2.get_positions(); print(ri)
```
