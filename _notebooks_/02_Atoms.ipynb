### Part 02.A <i><b>ase.atoms </b> --- molecules </i>

```python
from ase import Atoms
h2 = Atoms('H2', positions=[[0,0,0],[0,0,0.74]]); print(h2)
```

```python
zi = h2.get_atomic_numbers(); print(zi) # --- atomic number
ai = h2.get_chemical_symbols(); print(ai) # --- atomic symbol
ri = h2.get_positions(); print(ri) # --- atomic position
qi = h2.get_masses(); print(qi) # --- atomic mass
```

#### <i><b>set_positions</b></i>

```python
h2.set_positions([(0.,0.,0.),(0.,0.,-0.74)])
ri = h2.get_positions(); print(ri)
```

#### <i><b>translate</b></i>

```python
h2.translate([1.,0.,0.]) # --- translate 
ri = h2.get_positions(); print(ri)
```

#### <i><b>rotate</b></i>

```python
h2.rotate(90,'x',center='COM') # --- rotate 
ri = h2.get_positions(); print(ri)
```

### Part 02.B <i><b>ase.atoms</b> --- pbc </i>

```python
d = 3.0
L = 10.0
au = Atoms('Au',
    positions=[[0,0,0]],
    cell=[d, L, L],
    pbc=[1, 0, 0]
)

# --- pbc=[1,0,0] ???
# --- long wire of Au atoms
# x-direction => periodic
# y-direction => non-periodic
# z-direction => non-periodic
```

```python
zi = au.get_atomic_numbers(); print(zi) # --- atomic number
ai = au.get_chemical_symbols(); print(ai) # --- atomic symbol
ri = au.get_positions(); print(ri) # --- atomic position
qi = au.get_masses(); print(qi) # --- atomic mass
```

```python
pbc = au.get_pbc(); print(pbc) # --- periodic boundary condition flags
abc = au.get_cell(); print(abc) # --- unit cell vectors 
```

#### <i><b>set_pbc()</b></i>

```python
au.set_pbc([True, True, False])

pbc = au.get_pbc(); print(pbc)
```

#### <i><b>set_cell()</b></i>

```python
au.set_cell([3.0,3.0,10.0])

abc = au.get_cell(); print(abc)
```
