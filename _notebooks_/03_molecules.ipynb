## <i><b> from ase.build import molecule</b></i>

```python
from ase.build import molecule 
h2o = molecule('H2O')
```

```python
ai = h2o.get_chemical_symbols(); print(ai)
ri = h2o.get_positions(); print(ri)
```

Note: Only molecules that exist in the ASE molecule library will work. <br>
Example, C<sub>7</sub>H<sub>16</sub> molecule is NOT present in the library. <br>
Solution: SMILES => RDKit/Obabel (initial geometry) => ASE  

#### <i><b>ase.io</b></i>

```python
from ase.io import read,write
```

```python
mol = read('../_files_/phenol.xyz') # --- read the xyz file
ai = mol.get_chemical_symbols(); print(ai)
```

#### <i><b>get_distance()</b></i>

```python
dist = mol.get_distance(6,12); print(dist) # O-H bond length
```

#### <i><b>get_all_distances()</b></i>

```python
dists = mol.get_all_distances(); print(np.round(dists, 2))
```

#### <i><b>get_angle()</b></i>

```python
theta = mol.get_angle(5,6,12); print(theta)
```

#### <i><b>get_dihedral()</b></i>

```python
theta = mol.get_dihedral(4,5,6,12); print(theta)
```

#### <i><b>set_dihedral()</b></i>

```python
for dihedral in [0,90,120]:

    new_mol = mol.copy() # --- Shallow Copy
    new_mol.set_dihedral(4,5,6,12,dihedral)
    
    write(f"../_files_/phenol_dihedral_{dihedral:03d}.xyz", new_mol) # --- Save 
```
