```python
from ase import Atoms, units
from ase.calculators.emt import EMT
from ase.md.langevin import Langevin
from ase.md.velocitydistribution import MaxwellBoltzmannDistribution, Stationary
from ase.io.trajectory import Trajectory
```

```python


# --- 1. Create Atoms object (3 Cu atoms)
atoms = Atoms('Cu3',
              positions=[[0, 0, 0],
                         [2.5, 0, 0],
                         [0, 2.5, 0]])

# --- 2. Attach calculator (gives forces)
atoms.calc = EMT()

# --- 3. Initialize velocities (300 K)
MaxwellBoltzmannDistribution(atoms, temperature_K=300)
Stationary(atoms)  # remove net momentum

# --- 4. Set up MD (Langevin thermostat)
dyn = Langevin(atoms,
               timestep=1.0 * units.fs,
               temperature_K=300,
               friction=0.02)

# --- 5. Save trajectory
traj = Trajectory('md.traj', 'w', atoms)
dyn.attach(traj.write, interval=10)

# --- 6. Print info during run
def print_status():
    print(f"Step: {dyn.nsteps}, "
          f"T = {atoms.get_temperature():.2f} K, "
          f"E = {atoms.get_total_energy():.4f} eV")

dyn.attach(print_status, interval=10)

# --- 7. Run MD
dyn.run(100)
```
